# Steepest Descent
## Lecture 11

```python
import numpy as np
from sympy import Array
from sympy import Symbol, symbols
from sympy import derive_by_array

from steepest_descent import steepest_descent

# f1 IS FUNCTION FROM LECTURE 11, PAGE 11
# f2 IS FUNCTION FROM LECTURE 11, PAGE 14
# f1(x1, x2) = (x1 ** 2) + (25 * x2 ** 2)
# f2(x1, x2) = (x1 ** 2) + (2 * x1 * x2) + (2 * x2 ** 2) + x1
# FOR MORE INFO ABOUTE THE FUNCTIONS IMPORTED FROM steepest_descent.py use:
# print([function_name].__doc__) e.g. print(f1.__doc__)
from steepest_descent import f1, f2

%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
```

# EXAMPLE 1, LECTURE 11, PAGE 11

```python
# FUNCTION VARIABLES
x1, x2 = symbols('x1 x2')
#LEARNING RATE
alpha = 0.01
# STOP CONDITION THRESHOLD
threshold = 0.001
# INITIAL POINT
X0 = [0.5, 0.5]
# MAXIMUM NUMBER OF ITERATIONS
max_iter = 300
```

## INITIAL POINT: [0.5, 0.5], LEARNING RATE: 0.01

```python
X = steepest_descent(f1, alpha, max_iter, threshold, X0, x1, x2)
print(f'Optimum point: {X[-1]}')
print('5 first points:')
X[:5]
```

## CONTOUR PLOT

```python
x1_coordinates = np.linspace(-2, 2, 90)
x2_coordinates = np.linspace(-2, 2, 90)
X1, X2 = np.meshgrid(x1_coordinates, x2_coordinates)

Y1 = f1(X1, X2)

plt.figure(figsize=(10, 10))
plt.contour(X1, X2, Y1, levels=range(10), colors='gray')

plt.scatter([x[0] for x in X], [x[1] for x in X], marker='.', color='blue')
plt.plot([x[0] for x in X], [x[1] for x in X], color='r')
plt.title('Contour plot demo')
plt.xlabel('x1')
plt.ylabel('x2')
```

## CHANGING LEARNING RATE TO 0.03

```python
alpha = 0.03
X = steepest_descent(f1, alpha, max_iter, threshold, X0, x1, x2)
print(f'Optimum point: {X[-1]}')
print('5 first points:')
X[:5]
```

```python
Y1 = f1(X1, X2)

plt.figure(figsize=(10, 10))
plt.contour(X1, X2, Y1, levels=range(10), colors='gray')

plt.scatter([x[0] for x in X], [x[1] for x in X], marker='.', color='blue')
plt.plot([x[0] for x in X], [x[1] for x in X], color='r')
plt.title('Contour plot demo')
plt.xlabel('x1')
plt.ylabel('x2')
```

# STABLE LEARNING RATE (QUADRATIC)
## LECTURE 11, PAGE 16

```python
from steepest_descent import compute_quadratic_function_biggest_stable_learning_rate
print(compute_quadratic_function_biggest_stable_learning_rate.__doc__)
```

## EXAMPLE 1, LECTURE 11, PAGE 11, WITH THE BEST LEARNING RATE

```python
alpha = compute_quadratic_function_biggest_stable_learning_rate(f1, 0.001, x1, x2)
print(alpha)
```

## CONTOUR PLOT WITH LEARNING RATE 0.039

```python
X = steepest_descent(f1, alpha, max_iter, threshold, X0, x1, x2)

plt.figure(figsize=(10, 10))
plt.contour(X1, X2, Y1, levels=range(10), colors='gray')

plt.scatter([x[0] for x in X], [x[1] for x in X], marker='.', color='blue')
plt.plot([x[0] for x in X], [x[1] for x in X], color='r')
plt.title('Contour plot demo')
plt.xlabel('x1')
plt.ylabel('x2')
```

## ALGORITHM WIL NOT CONVERGE FOR BIGGER LEARNING RATES

```python
# AN INCREASE IN THE LEARNING RATE CAUSES THE DIVERGENCE OF THE ALGORITHM  
alpha += 0.001
X = steepest_descent(f1, alpha, max_iter, threshold, X0, x1, x2)

plt.figure(figsize=(10, 10))
plt.contour(X1, X2, Y1, levels=range(10), colors='gray')

plt.scatter([x[0] for x in X], [x[1] for x in X], marker='.', color='blue')
plt.plot([x[0] for x in X], [x[1] for x in X], color='r')
plt.title('Contour plot demo')
plt.xlabel('x1')
plt.ylabel('x2')
```
